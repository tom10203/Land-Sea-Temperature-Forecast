# Land and Sea Temperature Linear Regression Prediction Using SK-Learn

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib as plt
```

### Download the dataset and take a look at it

```python
df = pd.read_csv('Land.csv')
```

```python
df.head()
```

### Isolate the x and y values

```python
X = df['year'].reset_index()

y = df['Annual_mean'].values
```

### Here I used the pandas scatter graph module to plot the data in a scatter matirx. This gives me an initial visualisation as the wheather or not the is a correlation in the data.


```python
pd.plotting.scatter_matrix(df)
```

### Here I am going to create a model and fit the data to it. I can then plot a scatter graph and put a line of best fit through it to highlight the correlation.

```python
model = LinearRegression()
model.fit(X, y)
df["y_predict"] = model.predict(X)

a = df.plot.scatter(x="year", y="Annual_mean")
df.plot(x="year", y="y_predict", ax=a, color="Red")
```

### After fitting the model, I can then input future dates in to it and get it to predict future temperatures

```python
future = [i for i in range(2021, 2061)]
future = pd.Series(future).reset_index()

print(future.shape)
future_predict = model.predict(future)

predictions = pd.DataFrame()
predictions["future_year"] = future[0]
predictions["future_predict"] = future_predict
predictions.plot(x="future_year", y="future_predict")
```

### This is a crude way of predicting how much temperatures will rise in the future, however It does give a rough estimation and shows that there is a trend going upwards. To further increase the accuracy of the model I think I would need to gather more data and information on what is making the temperatures rise.
